#### 4) Qual variável é mais poderosa?

Considere as variáveis ```possui_email``` e ```posse_de_veiculo```. Faça uma tabela cruzada entre elas e responda qual delas te parece mais poderosa para prever a probabilidade de ```mau = 1```?

````python
df.posse_de_veiculo.unique()
````

````python
crosstab = pd.crosstab([df['possui_email'], df['posse_de_veiculo']], df['mau'], normalize='index')

crosstab = crosstab * 100

crosstab
````

Notemos que os os bons pagadores que possuem veiculo ou nao possuem uma diferenca de pouco mais de 0.3% para os que nao possuem e-mail; por outro lado os que possuem e-mail demonstram uma diferenca de 0.8% para os bons pagadores que possuem ou nao veiculo. Agora para os maus pagadores, a diferenca entre os que possuem veiculo ou nao eh de pouco mais de 0.3%; por outro lado para os que possuem e-mail, a diferenca entre os que possuem veiculo ou nao eh de pouco mais de 0.8%.

Com essa diferenca minima, podemos definir que nao ha significancia em escolher a posse de veiculo em detrimento da posse de e-mail, mostrando que nenhuma das duas eh mais forte que a outra.
